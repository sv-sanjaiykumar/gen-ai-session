```python
#See Special Token CLS and SEP
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
input_ids = tokenizer.encode(text, add_special_tokens=True)
print(input_ids)
```

```python
#Convert tokens to input ids
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
inputs = tokenizer(text)
print(inputs["input_ids"])
```

```python
#convert input ids back to tokens
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
input_ids = [101, 7592, 1010, 2129, 2024, 2017, 1029, 102]
tokens = tokenizer.convert_ids_to_tokens(input_ids)
print(tokens)
```

```python
#Load BERT model
from transformers import AutoModel
model = AutoModel.from_pretrained("bert-base-uncased")
print("BERT model loaded")
```

```python
#Generate embeddings
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
#view embedding in one tokens
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
embeddings = outputs.last_hidden_state[0][1]
print(embeddings.shape)
```

```python
#see actual embedding values
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
text = "Hello"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
embeddings = outputs.last_hidden_state[0][1]
print(embeddings[:5])
```

```python
#understand attention mask
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
inputs = tokenizer(text)
print(inputs["attention_mask"])
```

```python
#two-sentence input for NSP understanding
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text1 = "Hello, how are you?"
text2 = "hi there"
input = tokenizer(text1, text2)
tokens = tokenizer.convert_ids_to_tokens(input["input_ids"])
print(tokens)
```

```python
#extract cls embedding
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
text = "Hello, how are you?"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
cls_embedding = outputs.last_hidden_state[:,0,:]
print(cls_embedding)
```
